```python
import pandas as pd
import sweetviz
```

# Application Record

```python
app_record = pd.read_csv('application_record.csv')
```

```python
app_record_profiling = sweetviz.analyze(app_record)
```

```python
app_record_profiling.show_html('app_record_profiling.html', open_browser=True)
```

```python
app_record_profiling.show_notebook()
```

# Credit Record

```python
credit_record = pd.read_csv('credit_record.csv')
```

```python
credit_record_profiling = sweetviz.analyze(credit_record)
```

```python
credit_record_profiling.show_html('credit_record_profiling.html', open_browser=True)
```

```python
credit_record_profiling.show_notebook(w=1500, h=300, scale=0.8)
```

```python
status_categories = {'C' : 'Good', 'X' : 'Good', '0' : 'Good', '1' : 'Neutral', '2' : 'Neutral', '3' : 'Bad', 
                     '4' : 'Bad', '5' : 'Bad'}
credit_record.replace({'STATUS' : status_categories}, inplace=True)
```

```python
credit_record.STATUS.value_counts()
```

```python
merged = app_record.merge(credit_record, how='inner', on=['ID'])
```

```python
merged_profiling = sweetviz.analyze(merged)
```

```python
merged_profiling.show_notebook(w=1500, h=300, scale=0.8)
```
